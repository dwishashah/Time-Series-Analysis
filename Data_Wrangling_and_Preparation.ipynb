## Aggregation


```python
data.agg({'Smart_People': ['count','min', 'max', 'mean']})
```

```python
Query_string = """ select count(Smart_People) as count,
min(Smart_People) as min, max(Smart_People) as max,
avg(Smart_People) as mean from dfpsql;"""
sqldf(Query_string, globals())
```

## GROUP BY

```python
data.groupby('Country')['Smart_People'].sum()
```

```python
Query_string = """ select Country , sum(Smart_People) as Sum_Smart_People from data
group by Country;"""
sqldf(Query_string, globals())
```

## GROUP BY with Aggregation:

```python
data.groupby('SmartCity_Index_relative_Edmonton').agg({'Smart_People': ['mean','min','max']})
```

```python
Query_string = """ select SmartCity_Index_relative_Edmonton , avg(Smart_People) as mean, min(Smart_People) as
min, max(Smart_People) as max from data
group by SmartCity_Index_relative_Edmonton;"""
sqldf(Query_string, globals())
```

## JOIN MERGE

```python
import pandas as pd
data1 = {
'Empid': [1011, 1012, 1013, 1014, 1015],
'Name': ['John', 'Rahul', 'Rick', 'Morty', 'Tim'],
'Designation': ['Manager', 'Research Engineer', ' Research Engineer',
'VP', 'Delivery Manager'],'Date_of_joining': ['01-Jan-2000', '23-sep-2006', '11-Jan-2012', '21-Jan-1991', '12-Jan-1990']}
Emp_df = pd.DataFrame(data1, columns = ['Empid', 'Name',
'Designation','Date_of_joining'])
Emp_df
```

```python
data2 = {
'Empid': [1011, 1017, 1013, 1019, 1015],
'Deptartment': ['Management', 'Research', 'Research', 'Management',
'Delivery'],
'Total_Experience': [18, 10, 10, 28, 22]}
Dept_df = pd.DataFrame(data2, columns = ['Empid', 'Deptartment',
'Total_Experience'])
Dept_df
```

## INNER JOIN:

```python
pd.merge(Emp_df, Dept_df, left_on='Empid',right_on='Empid', how='inner')
```

```python
Query_string = """ select * from Emp_df a INNER JOIN Dept_df b ON a.Empid =
b.Empid;"""
sqldf(Query_string, globals())
```

## LEFT JOIN:

```python
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='left')
```

```python
Query_string = """ select * from Emp_df a LEFT JOIN Dept_df b ON a.Empid =
b.Empid;"""
sqldf(Query_string, globals())
```

## RIGHT JOIN:

```python
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='right')
```

```python
Query_string = """ select
a.Empid,Name,Designation,Date_of_joining,Deptartment,Total_Experience from
Dept_df a LEFT JOIN Emp_df b ON a.Empid = b.Empid;"""
sqldf(Query_string, globals())
```

## OUTER JOIN:

```python
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='outer')
```

```python
Query_string = """ select * from Emp_df a left OUTER JOIN Dept_df b
ON a.Empid = b.Empid;"""
sqldf(Query_string, globals())
```
